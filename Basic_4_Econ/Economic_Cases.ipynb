# **Simple Economic Model**
**Scenario:**

Let's build a basic supply and demand model. We'll assume linear supply and demand curves and calculate the equilibrium price and quantity.

**Supply:** The quantity of a good or service that producers are willing and able to offer for sale at various prices.

**Demand:** The quantity of a good or service that consumers are willing and able to buy at various prices.

**Equilibrium:** The point where the supply and demand curves intersect, representing the price at which the quantity demanded equals the quantity supplied.

**Steps:**
1. Import necessary libraries: numpy for numerical operations and matplotlib for plotting.
2. Define supply and demand functions: Linear functions with parameters.
3. Find equilibrium: Uses a bisection method to find the price where supply equals demand.
4. Generate data for plotting: Creates price and quantity data for both supply and demand curves.
5. Plot the curves: Visualizes the supply and demand curves.
6. Print equilibrium values: Displays the calculated equilibrium price and quantity.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def supply_function(price):
  """Linear supply function."""
  a = 10  # Intercept
  b = 2  # Slope
  quantity = a + b * price
  return quantity
```

```python
def demand_function(price):
  """Linear demand function."""
  c = 100  # Intercept
  d = -3  # Slope
  quantity = c + d * price
  return quantity
```

```python
def find_equilibrium(supply_func, demand_func):
  """Finds the equilibrium price and quantity."""
  low_price = 0
  high_price = 100
  tolerance = 0.001
  while high_price - low_price > tolerance:
    mid_price = (low_price + high_price) / 2
    supply_quantity = supply_func(mid_price)
    demand_quantity = demand_func(mid_price)

    if supply_quantity > demand_quantity:
      high_price = mid_price
    else:
      low_price = mid_price

  equilibrium_price = mid_price
  equilibrium_quantity = supply_func(equilibrium_price)
  return equilibrium_price, equilibrium_quantity
```

```python
# Example usage
price_range = np.linspace(0, 20, 100)  # random integer
supply_quantities = [supply_function(price) for price in price_range]
demand_quantities = [demand_function(price) for price in price_range]

equilibrium_price, equilibrium_quantity = find_equilibrium(supply_function, demand_function)
```

```python
plt.plot(price_range, supply_quantities, label='Supply')
plt.plot(price_range, demand_quantities, label='Demand')
plt.xlabel('Price')
plt.ylabel('Quantity')
plt.title('Supply and Demand')
plt.scatter(equilibrium_price, equilibrium_quantity, color='red', label='Equilibrium')
plt.legend()
plt.show()

print(f"Equilibrium price: {equilibrium_price}")
print(f"Equilibrium quantity: {equilibrium_quantity:.3f}")
```

# **Exercise Study: Price Elasticity of Demand**

Price elasticity of demand measures the responsiveness of quantity demanded to changes in price. It is calculated as the percentage change in quantity demanded divided by the percentage change in price.


**Dataset Structure**
A basic dataset for this analysis might include:

- product_id: Unique identifier for each product

- price: Price of the product

- quantity_sold: Quantity sold at the given price

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Sample data
data = {
  'product_id': [1, 1, 1, 1],
  'price': [10, 12, 8, 9],
  'quantity_sold': [100, 90, 110, 105]
}
df = pd.DataFrame(data)
```

```python
# Create a scatter plot to visualize the relationship between price and quantity sold
plt.scatter(df['price'], df['quantity_sold'])
plt.xlabel('Price')
plt.ylabel('Quantity Sold')
plt.title('Price vs Quantity Sold')
plt.show()
```
