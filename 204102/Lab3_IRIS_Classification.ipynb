อ่านข้อมูลจากไฟล์

```python
# Load libraries
import pandas as pd
cols = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Class']
dataset = pd.read_csv("iris.csv",names=cols)
dataset.head()
```

แบ่งข้อมูลเป็น Features และ Class

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 4].values.reshape(-1,1);y
# X = ค่าที่จะเอาไปทำนาย -> attribute
# y = คำตอบ -> Class
```

ทำการแบ่ง Training set และ Testing set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.60, stratify=y) # stratify = แบ่งแบบภานในการสุ่ม train และ test มีการแบ่ง class ให้ balanced
                                                                                      # เช่น จากข้อมูลทั้งหมด 150 และมีการแบ่ง test 30 ชุด และมี class 3 คลาส
                                                                                      # ดังนั้นจะเกิดการแบ่งข้อมูลใน train 10 ชุด เป็น class ที่ 1
                                                                                      # train อีก 10 ชุด เป็น class ที่ ​​2
                                                                                      # และtrain 10 ชุดสุดท้าย เป็น class ที่ ​​3
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

สร้างโมเดลการจำแนกข้อมูล

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X_train,y_train) # train model เกิดขึ้นที่ .fit
```

การทำนายผลจากโมเดลที่สร้างขึ้น

```python
y_pred = model.predict(X_test) 
# เอาข้อมูล ที่ใช้ test ส่งเข้าไปให้ model ที่ผ่านการ train 
# แล้วทายผล แล้วก็นำมาวัดค่า Accuracy score เพ่ือหาค่าความแม่นยำ
print(y_pred)
```

ประเมินผลการทดสอบด้วยค่า Accuracy

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred) 
# y_test คือ ผลเฉลย นำมาเทียบกับ y_pred ซึ่งเป็นค่าพยากรณ์
print(accuracy) # ค่าจะสุ่มเรื่อย ๆ ตาม y_pred ที่ถูกนำมาสุ่ม
```

ประเมินผลการทดสอบด้วย Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred) # confusion matrix คือ

"""               Iris-setosa           Iris-versicolor         Iris-virginica

Iris-setosa           30                       0                      0

Iris-versicolor        0                      29                      1

Iris-virginica         0                       3                     27
"""

print(cm)
```

```python
from sklearn.metrics import classification_report
print (classification_report(y_test, y_pred))
```

นำเข้า Libaries สำหรับการจำแนกแบบอื่น

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

ทำการสร้างและทดสอบโมเดลด้วย classifier algorithms อื่น

```python
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

models = []                                                      # model เป็น array ที่เก็บได้หลาย model
models.append(('KNN-5',   KNeighborsClassifier()              )) # n_neightbors (dafault) = 5
models.append(('KNN-10',  KNeighborsClassifier(n_neighbors=10)))
models.append(('Tree',    DecisionTreeClassifier()            ))

# evaluate each model in turn
# หา model ที่ดีที่สุด แล้วเอา ค่า
  # round 1 -> name = KNN-5 , clf = KNeighborsClassifier()
  # round 2 -> name = KNN-10, clf = KNeighborsClassifier(n_neightbors=10)
  # round 3 -> name = Tree  , clf = DecisionTreeClassifier()
results = []
names = []
scoring = 'accuracy'
for name, clf in models:
  # Create each Model
  model = clf
  # Train Model
  model.fit(X_train,y_train)
  # Predict Model
  Predited_class = model.predict(X_test)
  # Finds Accuracy Score
  accuracy = accuracy_score(y_test, Predited_class)
  results.append(accuracy)
  names.append(name)
  msg = "%s: %f " % (name, accuracy)
  print(msg)


# boxplot algorithm comparison
fig = plt.figure()
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
plt.bar(names,results)

plt.show()
```
