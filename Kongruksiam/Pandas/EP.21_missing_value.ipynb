```python
import pandas as pd
import numpy as np
import os 
os.chdir("/Users/phutharesuanwachirapisut/Programming/Dataset")
df = pd.read_excel("Employee.xlsx",index_col="Name");df
# df.drop(["binart_class"],axis=1,inplace=True)
# df.drop(["binary_class"],axis=1,inplace=True)
```

```python
# Missing Value
    # df.isnull()
    # df.notnull()
    # df.isnull().any() -> True means this df has at least one missing value
    # df.isnull().sum()
    # df.notnull().any()
    # df.notnull().sum()
	# any() เพื่อตรวจสอบว่ามีค่า True อย่างน้อยหนึ่งค่าใน DataFrame หรือ Series หรือไม่ 
df
```

```python
# แทน missing value ด้วย avg

df.describe()
select = "Salary"
df[select] = df[select].fillna(df[select].mean()) # เหมือน r programming
# column.fillna(column.ค่าที่จะแทน)
df
```

```python
# แทนค่าข้อมูลด้วยค่าที่กำหนดขึ้นมา

select = "Bonus"
df[select] = df[select].fillna(0.1);df
```

```python
# แทนค่าด้วยค่าก่อนหน้า (row ก่อนหน้า)
    # ถ้าไม่กำหนด column จะหมายถึงทั้ง Table
a= df.fillna(method="pad");a
```

```python
# แทนค่าด้วยค่าถัดไป (row ถัดไป)

df.fillna(method="bfill",inplace=True);df
```

```python
# delete row
df.dropna(subset=["Age","Job"],inplace=True);df
# เงื่อนไขว่าถ้าเป็น NaN ที่ column Age กับ Job จะลบทิ้ง 
# ถ้าไม่ใส่ subset จะลบ row โดยการดูจากทุก column
```

```python
# delete column
df.dropna(axis="columns",inplace=True);df
# ลบทุก column ที่มี NaN
```

```python
# การลบค่าซ้ำ

df[df.duplicated()]
df.drop_duplicates(inplace=True)
df
# จากตัวอย่าง มีการลบ เกษมสันต์
# 
```
