```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

# Machine Learning แบบมีผู้สอน หรือก็คือมีเฉลย หรือเรียกว่า class
```

Naive Bayes

    1. แยกพิจารณาตาม class
    2. หา Prop. ของแต่ละ attribute
    3. หา Prop. ของ class ตามที่เราแยกออกมา
    4. เอาทุกค่ามาคูณกัน

ใข้ Gaussian

```python
# Load Dataset
iris = load_iris()

# Assign Attribute and target
x = iris["data"]
y = iris["target"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4)
```

```python
# Model
model = GaussianNB()

# Train Model
model.fit(x_train, y_train)
```

```python
# Prediction
y_pred = model.predict(x_test)
# Accuracy Score
print(f"Accuracu Score : {accuracy_score(y_test, y_pred)}")
```

ทำนายรายได้ประชากร ด้วย Gaussian Naive Bayes

```python
# Import Dataset
adult_df = pd.read_csv("/Users/phutharesuanwachirapisut/Downloads/Programming/Dataset/adult.csv")
# print(adult_df.describe)
print(adult_df.head()) # -> found missing value
```

```python
def cleandata(dataset):
    for column in dataset.columns:
        if dataset[column].dtype == type(object):
            le = LabelEncoder()
            dataset[column] = le.fit_transform(dataset[column])
    return dataset

def split_feature_class(dataset, feature):
    features = dataset.drop(feature, axis=1)    # del income column
    labels = dataset[feature].copy()            # select only income column
    return features,labels
```

```python
# Cleaning Data
adult_df = cleandata(adult_df)
print(adult_df.head())
```

```python
# แบ่งข้อมูล เพื่อ
# 1st Split adult_df
training_set, test_set = train_test_split(adult_df, test_size=0.2)

# 2nd Split training_set
train_features, train_labels = split_feature_class(training_set,"income")

# 2nd Splite test_set
test_features, test_labels = split_feature_class(test_set, "income")
```

```python
# Model
model = GaussianNB()
model.fit(train_features,train_labels)
```

```python
# Prediction
clf_pred = model.predict(test_features)
```

```python
# Accuracy Score
print(f'Accuracy = {accuracy_score(test_labels,clf_pred)}')
```
