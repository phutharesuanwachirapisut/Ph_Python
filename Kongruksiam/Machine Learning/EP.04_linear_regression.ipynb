```python
# y = ax + b
import numpy as np
import matplotlib.pyplot as plt

# random 100 float numbers between -5 to 5 
x = np.linspace(-5,5,100) # no lin "e" space() but linspace()
print(x)
```

```python
y = (2 * x**1) + 1
plt.plot(x,y,"-r",label="y=2x+1")

plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc="upper left")
plt.title("Graph y=2x+1")
plt.grid()
plt.show()
```

Linear regression Model
    1. จำลองข้อมูล
    2. Linear Regression Model
        2.1 Train Data Algorithm
    3. Test Model
        ได้ผลลัพธ์จากการคาดการณ์
    4. Analysis Model 

```python
from sklearn.linear_model import LinearRegression
# 1. จำลองข้อมูล

rng = np.random 
x = rng.rand(50) * 10 # สุ่มได้เฉพาะค่าบวก
c = rng.randn(50) # สุ่มได้ค่าลบด้วย
y = (2 * x) + c
```

```python
# 2. Linear Regression Model
model = LinearRegression()
x_new = x.reshape(-1,1)
    # 2.1 Training the data algorithm ** Only array2D
model.fit(x_new,y)
print(model.score(x_new, y)) # show R-SQuare 
```

```python
# 3. Test Model
xfit = np.linspace(-1,11).reshape(-1,1)
yfit = model.predict(xfit)
```

```python
# 4. Analysis Model
plt.scatter(x,y)
plt.plot(xfit,yfit)
plt.show()
```
