```python
import matplotlib.pyplot as plt
import numpy as np

course = ["PHP","java",'python',"C++","SQL"]
students = [15,30,50,45,20]

color = ["gray","red","blue","green","purple"]
```

```python
plt.bar(
    course,
    students,
    color=color,
    width=0.6,          # ความหนาของ bar
    edgecolor="red",    # สีเส้นขอบ
    linewidth=3         # ขนาดเส้นขอบ
    )
plt.xlabel("Course",size=15)
plt.ylabel("Students",size=15)
plt.title("Total Students",size=17)
plt.show()
```

```python
plt.barh(
    course,
    students,
    color=color,
    # width=0.6,          # ความหนาของ bar 
        # plt.barh() ไม่มี width
    edgecolor="red",    # สีเส้นขอบ
    linewidth=3         # ขนาดเส้นขอบ
    )
plt.ylabel("Course",size=15) # สลับ bar เป็น barh แล้วอย่างลทมเปลี่ยนชื่อแกน ชื่อแกนไม่เปลี่ยนตามให้
plt.xlabel("Students",size=15)
plt.title("Total Students",size=17)
plt.show()
```
