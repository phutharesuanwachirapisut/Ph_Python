```python
import matplotlib.pyplot as plt
import numpy as np

arr_a = np.arange(1,5)
products_1 = [30,10,20,60,40]
products_2 = [15,20,30,50,66]
months = list(range(1,6))
plt.plot(months,products_1,"go-")
plt.plot(months,products_2,"r*-")

plt.xlabel(
    "Months",
    size=10,
    color="c")

plt.ylabel(
    "Sales",
    size=10,
    color="m")

config = {
    "size":15,
    "color":"w",
    "backgroundcolor":"k"}

plt.title(
    "Top Sale 2024",
    config,
    loc="left") # loc = [left,right,center(default)]

plt.text(2,10,"pointA",size="xx-large",color="m",backgroundcolor="k") # ตำแหน่งที่ (2,10) จะแสดงชื่อข้อความว่า "pointA"
plt.text(1.5,60,"pointB",config)
plt.show()

```
