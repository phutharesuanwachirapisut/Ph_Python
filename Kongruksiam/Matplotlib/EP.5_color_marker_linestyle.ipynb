```python
import matplotlib.pyplot as plt
products_1 = [1.2, 2.8, 1.3, 4.2, 3.5]
products_2 = [5.5, 10.0, 3.6, 1.0, 2.2]
months = list(range(1,6))

plt.plot(months,products_1)
plt.plot(months,products_2)
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
'''
b = blue
g = green
r = red
c = cyan
m = magenta
y = yellow
k = black
w = white
'''
plt.plot(months,products_1,color="m")
plt.plot(months,products_2,color="k")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

```python
'''
. = point
o = cycle
x = x marker
+ = plus
* = star
s = square
| = vline
_ = hline
v = down
^ = up
> = right
< = left
'''
plt.plot(months,products_2,color="m",marker="o")
plt.show()
```

```python
'''
- = solid 
-- = dahsed 
-. = dash-dot 
: = dotted 
'''
plt.plot(months,products_2,color="m",marker="o",linestyle=":")
plt.show()
```

```python
# Foramt String
# '[marker][line][color]'
# '[color][marker][line]'

plt.plot(months,products_2,"ms--")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```
