```python
import matplotlib.pyplot as plt
products_1 = [1.2, 2.8, 1.3, 4.2, 3.5]
products_2 = [5.5, 10.0, 3.6, 1.0, 2.2]
months = list(range(1,6))

plt.plot(months,products_1)
plt.plot(months,products_2)
plt.xlabel("Months")
plt.ylabel("Sales")
# save file
plt.savefig("export1.png",transparent=True)
    # trasparent = True คือการทำให้เป็นภาพโปร่ง แบบ png
# save file
plt.show()
```
