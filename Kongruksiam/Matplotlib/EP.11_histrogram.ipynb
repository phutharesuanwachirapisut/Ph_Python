```python
import matplotlib.pyplot as plt
import numpy as np
import random

# Histrogrma แบบไม่ต่อเนื่อง 
    # สั่งเกตว่ากราฟจะมีรู
age = []

for i in range(16):
    age.append(random.randint(0,100))
print("list age = {}".format(age))
```

```python
plt.hist(age)
plt.xlabel("Age",size=15,color="m")
plt.ylabel("Count",size=15,color="m")
plt.title("Total Age",size=17,color="m")
plt.show()
```

```python
data=np.random.normal(10,2.5,5000) # สุ่มข้อมูล 5000 ชุด ค่าเฉลี่ย 10 ส่วนเบี่งเบนมาตรฐาน 2.5

plt.hist(data,bins = 30) # ช่วงข้อมูล 30 ช่วง
plt.xlabel("Data",size=15,color="m")
plt.ylabel("Count",size=15,color="m")
plt.title("Total Age",size=17,color="m")
plt.show()
```
