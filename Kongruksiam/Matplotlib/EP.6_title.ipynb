```python
import matplotlib.pyplot as plt
products_1 = [1.2, 2.8, 1.3, 4.2, 3.5]
products_2 = [5.5, 10.0, 3.6, 1.0, 2.2]
months = list(range(1,6))

plt.plot(months,products_1,"bo--")
plt.plot(months,products_2,"ro--")
plt.xlabel("Months",size=20,color="blue",backgroundcolor="red")
plt.ylabel("Sales",size=20,color="red",backgroundcolor= "blue")
# -----title-------
config = {"size":20,"color":"w","backgroundcolor":"k"}
plt.title("Top Sale 2020",config,loc="left")
# -----title-------

plt.show()
```
