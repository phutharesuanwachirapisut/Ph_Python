```python
import matplotlib.pyplot as plt
import numpy as np
students = [15,30,45,60,20]
course = ["PHP","java",'python',"C++","SQL"]
color = ["gray","red","blue","green","purple"]
exp = [0.13,0.22,0.25,0.3,0.05] # ค่าระยะ
```

```python
plt.pie(
    students,
    labels=course,      # จำเป็นต้องใส่ labels
    autopct="%.1f%%",   # แสดงเป็นทศนิยม 1 ตำแหน่ง
    colors=color,       # ตั้งค่าสีของตัวแฟรแต่ละชนิด
    shadow=True,        # ตามชื่อ คือ แสดงเงา เพื่อเพิ่มมิติ
    explode=exp,        # เป็นการดึงชิ้นส่วนออกมา ตามค่าที่เราระบุ
    startangle = 0)     # เป็นการหมุนการกราฟวงกลม
plt.show()
```
