```python
import matplotlib.pyplot as plt
import numpy as np
students_A = [15,30,45,60,20]
students_B = [10,40,20,24,60]

course = ["PHP","java",'python',"C++","SQL"]
color = ["gray","red","blue","green","purple"]
```

```python
plt.bar(
    course,
    students_A,
    label="Students_A",
    color="r"
)

plt.bar(
    course,
    students_B, 
    bottom=students_A,  #ให้ student_A อยู่ข้างล่าง
    label="Students_B",
    color="k"
)

plt.legend()
```

```python
plt.stackplot(
    course,
    students_A,
    students_B,
    labels=["Students_A","Students_B"], # กำหนดชื่อของพื้นที่ เรียงลำดับตามการเรียกตัวแปร
    colors=["r","k"])                   # กำหนดสี เรียงลำดับตามการเรียกตัวแปร
plt.legend()
```
