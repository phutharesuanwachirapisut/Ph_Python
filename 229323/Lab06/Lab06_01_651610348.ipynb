```python
import pandas as pd
import csv
```

```python
f = open("stu_grade.txt")

# class my_dialect(csv.Dialect):
#   lineterminator='\n'
#   delimiter=':'
#   quotechar='"'
#   quoting=csv.QUOTE_MINIMAL

reader=csv.reader(f,delimiter=":")

data = []

for line in reader:
	line = list(map(lambda x: x.strip() ,line))
	if len(line) == 1:
		line = line[0].split("=")
		line = list(map(lambda x: x.strip(),line))
	data.append(line)
data = data[:len(data) - 3];data

	
```

```python

dict_ = {"Name":[],"Program":[],"G.P.A.":[]}
for item in data:
	if dict_[item[0]] == []:
		dict_[item[0]] = [item[1]]
	else:
		if item[0] != "G.P.A.":
			dict_[item[0]] += [item[1]]
		else:
			item[1] = float(item[1])
			dict_[item[0]] += [float(item[1])]
df = pd.DataFrame(dict_)
df["G.P.A."] = df["G.P.A."].apply(pd.to_numeric)
```

```python
# 1)  จํานวนของประเภท Program ว่าแต่ละแบบมีนักเรียนทั้งหมด กี่คน และชื่ออะไรบ้าง   
df1 = df[["Name","Program"]]

df1 = df1.set_index(["Program","Name"]);df1
```

```python
df2 = df1.groupby(["Program"]).size().reset_index(name="Name")
df2 = df2.set_index(["Program","Name"]);df2
```

```python
df3 = df1.groupby(["Program","Name"]).count();df3
```

```python
result = pd.concat([df2,df3])
result = result.groupby(["Program","Name"]).count()
result
```

```python
result = pd.concat([df2,df3])
result = result.groupby(["Program","Name"]).count()

sort_dict = {"C":0, "Python":1 , "Java":2}
result = result.sort_values(by=["Program"],key= lambda x: x.map(sort_dict))
result
```

```python
# 2)  รายชื่อของนักเรียนที่มี G.P.A. อยู่ในช่วง 2-2.99 และ 3.00 ขึ้นไป
df10 = df[["G.P.A.","Name"]]
df10.loc[df10["G.P.A."] > 3]
df10.loc[(df10["G.P.A."] > 2) & (df10["G.P.A."] < 2.99)]
df10
```

```python
def grade(x):
	if x >= 3.00:
		return  ">=3.00"
	elif x >= 2.00 and x <= 2.99:
		return "2.00-2.99"

```

```python
df10["G"] = df10["G.P.A."].apply(grade)
df10
```

```python
result = df10[["G","Name"]]
result.set_index(["G","Name"])
result.groupby(["G","Name"]).count()
```
