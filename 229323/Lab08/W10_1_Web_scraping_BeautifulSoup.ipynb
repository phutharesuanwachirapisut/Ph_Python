# **การใช้ regular expression**

## หาแท็กที่ชื่อชนิดมีตัว h ตามด้วยตัวเลข ได้แก่พวก h3 h4

```python
import re

print(sup.find_all(re.compile(r'h\d')))
```

## หาแท็กที่มี id ขึ้นต้นด้วย w

```python
ww = sup.find_all(id=re.compile(r'w.+')) 	# w = ทุกตัวอักษร
											# . = ตัวอะไรก็ได้ 
											# . หลัง w คือ ตามด้วยตัวอะไรก็ได้
											# + = หลายครั้ง
print(ww)
```

```python
re.findall('so+','so soooooo s see sooo sb soso ssso scbo') # so+ => so อย่างน้อย 1 ตัว
```

```python
re.findall('so*','so soooooo s see sooo sb soso ssso scbo') # so* => ไม่มี so หรือมีก็ได้
```

```python
re.findall('s.+','so soooooo s see sooo sb soso ssso scbo') # s.+ => s ตามด้วยทุกตัวอักษรยกเว้นการขึ้นบรรทัดใหม่ อย่างน้อย 1 ตัว
```

```python
re.findall('<.+>','<a> xxx <bc> xxxx <def> xxx') # <.+> => < ตามด้วยทุกตัวอักษรยกเว้นการขึ้นบรรทัดใหม่ อย่างน้อย 1 ตัว และลงท้ายด้วย >
```

```python
re.findall('<.+?>','<a> xxx <bc> xxxx <def> xxx') # ใส่ ? จะเป็นการตัดเป็น item
```

## การใช้งานที่มักพบบ่อยก็คือใช้เพื่อค้นหา url ของรูปภาพที่ต้องการจากหน้าเว็บแล้วทำการโหลด

```python
url_naweb = 'https://hinaboshi.com/súpđẹp.html'
r = requests.get(url_naweb)
r.encoding = 'utf-8'
sup = BeautifulSoup(r.text,'lxml')

for klong in sup(class_='klonglek'):
    src = klong.img['src']
    chue_file = src.split('/')[-1]
    
    url_file = 'https://hinaboshi.com/'+src
    
    with open(chue_file,'wb') as f:
        f.write(requests.get(url_file).content)
```
