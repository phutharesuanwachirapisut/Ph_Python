## C
จากไฟล์ราคาหุ้นแต่ละตัว ในที่นี้จะมี 5 ตัว เราจะพิจารณาค่า column เพียง 5 ค่า ได้แก่ Open, High, Low, Close
และ Volume เท่านั้น จากนั้นทําการสร้าง dataFrame เพื่อเก็บข้อมูลสําหรับหุ้นแต่ละตัว โดยมี 5 columns ที่เก็บ 
Open, High, Close และ Volume ที่สูงที่สุดของแต่ละเดือนในช่วง 3 เดือน (ได้แก่เดือน ก.ค., ส.ค. และ ก.ย.) ขณะที่ 
column Low จะเก็บราคาที่ต่ําที่สุดของแต่ละเดือนในช่วง 3 เดือน 
กรณีมีข้อมูลว่างหรือ NaN ให้ข้ามค่าข้อมูลนั้นไปไม่ต้องนํามาเปรียบเทียบ

```python
list_a = [df_AH, df_BBL, df_PTT, df_SCG, df_TRUE]
list_m = ["Jul", "Aug", "Sep"]
list_comp = "AH, BBL, PTT, SCG, TRUE".split(", ")
count_ = 0
result_c = pd.DataFrame(columns="Open,High,Low,Close,Volume,Month,Stock".split(","))
for m in list_m:
	for bus in list_a:

		bus = bus.set_index(bus.index.get_level_values('Months'))
		bus = bus.drop(["Adj Close"], axis=1)
		
		bus_ = pd.DataFrame(bus.loc[(bus.index == m)].agg({
			'Open': 'max',
			'High': 'max',
			'Low': 'min',
			'Close': 'max',
			'Volume': 'max'
		})).T
		bus_["Stock"] = list_comp[count_%5]
		count_ +=1
		bus_["Month"] = m

		result_c = pd.concat([result_c, bus_])
		
		

```

```python
result_c = result_c.reindex("Month Stock Open High Low Close Volume".split(" "), axis=1)
result_c = result_c.set_index(result_c.Month).drop("Month",axis=1)
result_c
```

## E
สร้าง dataFrame เพื่อเก็บผลต่างระหว่างค่า Open และ Close คือคํานวณ (Open-Close) ของหุ้นแต่ละตัวในแต่ละ
วัน(ทั้ง 3 เดือน) กรณีไม่มีค่าใดค่าหนึ่ง ไม่มีค่า Open หรือค่า Low ให้เก็บค่าเป็น NaN โดยเรียงวันจาก 30 ก.ย.
2023 ไป 1 ก.ค. 2023

```python
df_AH["Open-Close"] = df_AH["Open"] - df_AH["Close"]
dff_ah = pd.DataFrame(df_AH.loc[:, "Open-Close"])
dff_ah = dff_ah.loc[(dff_ah.index.get_level_values("Months") == "Sep") | (dff_ah.index.get_level_values("Months") == "Aug") | (dff_ah.index.get_level_values("Months") == "Jul")]
dff_ah.reset_index(inplace=True)
dff_ah.set_index("Date", inplace=True)
dff_ah = dff_ah.drop("Months",axis=1).sort_index(ascending=False)
dff_ah
```

```python
df_BBL["Open-Close"] = df_BBL["Open"] - df_BBL["Close"]
dff_bbl = pd.DataFrame(df_BBL.loc[:, "Open-Close"])
dff_bbl = dff_bbl.loc[(dff_bbl.index.get_level_values("Months") == "Sep") | (dff_bbl.index.get_level_values("Months") == "Aug") | (dff_bbl.index.get_level_values("Months") == "Jul")]
dff_bbl.reset_index(inplace=True)
dff_bbl.set_index("Date", inplace=True)
dff_bbl = dff_bbl.drop("Months",axis=1).sort_index(ascending=False)
dff_bbl
```

```python
df_PTT["Open-Close"] = df_PTT["Open"] - df_PTT["Close"]
dff_ptt = pd.DataFrame(df_PTT.loc[:, "Open-Close"])
dff_ptt = dff_ptt.loc[(dff_ptt.index.get_level_values("Months") == "Sep") | (dff_ptt.index.get_level_values("Months") == "Aug") | (dff_ptt.index.get_level_values("Months") == "Jul")]
dff_ptt.reset_index(inplace=True)
dff_ptt.set_index("Date", inplace=True)
dff_ptt = dff_ptt.drop("Months",axis=1).sort_index(ascending=False)
dff_ptt
```

```python
df_SCG["Open-Close"] = df_SCG["Open"] - df_SCG["Close"]
dff_scg = pd.DataFrame(df_SCG.loc[:, "Open-Close"])
dff_scg = dff_scg.loc[(dff_scg.index.get_level_values("Months") == "Sep") | (dff_scg.index.get_level_values("Months") == "Aug") | (dff_scg.index.get_level_values("Months") == "Jul")]
dff_scg.reset_index(inplace=True)
dff_scg.set_index("Date", inplace=True)
dff_scg = dff_scg.drop("Months",axis=1).sort_index(ascending=False)
dff_scg
```
