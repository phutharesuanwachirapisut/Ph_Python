Creating a Dictionary

```python
#Empty dictionary
d={}
print(d)

#key:value
d={"cow":5,"dog":98,"cat":1}
print(d)

#ใช้ฟังก์ชัน dict()
d=dict()
print(d)

#สร้าง dictionary จาก list ของ Tuple(key,value)
pairs=[("one",1),("two",2),("three",3)]
d=dict(pairs)
print(d)

#สร้าง dict โดยใช้ function parameters
d=dict(one=1,two=2,three=3)
print(d)
```

Example Using Dictionaries

```python
#ฟังก์ชันสำหรับหาค่า Element ที่มีความถี่มากที่่สุุดใน list a
def most_frequent(a):
  max_value=None
  max_count=0
  frequency=dict()
  for element in a:
      if element in frequency:
        count=frequency[element]
      else:
        count=0
      count+=1
      frequency[element]=count
      if(count>max_count):
        max_count=count
        max_value=element
  return max_value

print(most_frequent([2,5,3,4,6,4,2,4,5]))
```

```python
#ฟังก์ชันสำหรับหาค่า Element ที่มีความถี่มากที่่สุุดใน list a version2
def most_frequent(a):
  max_value=None
  max_count=0
  frequency=dict()
  for element in a:
      count=1+frequency.get(element,0)
      frequency[element]=count
      if(count>max_count):
        max_count=count
        max_value=element
  return max_value

print(most_frequent([2,5,3,4,6,4,2,4,5]))
```
