zip and unzip

```python
s_id=['Math','Bio','Chem']
score1=[4.3,5.2,3.6]
zipped=zip(s_id,score1)
print("1",zipped)

subj_score1=list(zipped) #แปลง zipped ให้เป็น list
print("2",subj_score1)

score2=[3.2,1.6,4.8]
subj_score1_2=list(zip(s_id,score1,score2)) #เอา list 3 อันมา zip กัน แล้วแปลงให้เป็น list
print("3",subj_score1_2)

a,b,c=zip(*subj_score1_2) #unzip subj_score1_2  ผลลัพธ์ได้ออกมาเป็น 3 tuples
print("4 ","a=",a,"-->type(a)=",type(a))
print("5 ","b=",b,"-->type(b)=",type(b))
print("6 ", "c=",c,"-->type(c)=",type(c))





```

```python
def numbers_with_3s(lo,hi):
    result=[]
    for x in range(lo,hi):
      if ("3" in str(x)):
        result.append(x)
    return result

print(numbers_with_3s(250,304))
```

```python
def stdev(a):
  sum=0
  for x in a:
    sum+=x
  xbar=sum/len(a)
  print(sum)
  print(xbar)

  sum2=0
  for x in a:
    z=pow(x-xbar,2)
    sum2+=z
  print(sum2)

  value=sum2/len(a)
  print(value)

  #return(value**0.5)
  return(pow(value,0.5))



print(stdev([2.5,3,4.5,6.5,9,10]))
```

```python
def repeats(a):
  seen=set()
  seen_again=set()
  for element in a:
    if (element in seen):
      seen_again.add(element)
    seen.add(element)
  return sorted(list(seen_again))

print(repeats([2,5,3,4,6,4,2]))
```
