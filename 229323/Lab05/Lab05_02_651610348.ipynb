```python
import files
import pandas as pd
import io 
import os 
import csv
# uploaded=files.upload()
os.chdir("/Users/phutharesuanwachirapisut/Programming/Python/229323/Lab05")
```

```python
df = pd.read_csv("data_651610348_0501.csv");df
```

```python
def BMI_caculate(height, weight):
    return round(weight/(height**2),4)
```

```python
df["BMI Value"] = BMI_caculate(height=df["Height(metre)"], weight=df["Weight(kg.)"])
df
```

```python
def BMI_mean(value):    
    if value <= 18.5:
        return "Thin" 
    elif 18.6 <= value and value <= 24.9:
        return "Healthy" 
    elif 25 <= value and value <= 29.9:
        return "Overweight" 
    elif value >= 30:
        return "Obese" 

df["Mean"] = df["BMI Value"].apply(BMI_mean)
df
```

```python
data_list = list(df.itertuples(index=False,name=None))

result = open("data_651610348_0502.csv","w")
result = csv.writer(result)

result.writerow(["Name","Height(metre)","Weight(kg.)", "BMI Value", "Mean"])
for i in range(len(data_list)):
    result.writerow([data_list[i][0],data_list[i][1],data_list[i][2],data_list[i][3],data_list[i][4]])
```
