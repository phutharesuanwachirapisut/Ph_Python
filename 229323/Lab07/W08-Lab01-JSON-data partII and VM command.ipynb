การแปลง object เป็น json

คำสั่ง dumps
ถ้าข้อมูลเป็น dictionary ของ list และ tuple

```python
import io
import json
ab={'a':[1,2], "b":(3,4)}  # Obj
print(ab)
print(json.dumps(ab))  # สังเกตว่าเป็น dumps  ที่เติมตัว s ด้วย จะเป็นการแปลง object เป็นข้อความ json

# จุดสังเกต
# {'a': [1, 2], 'b': (3, 4)}
# {"a": [1, 2], "b": [3, 4]}
	# single quotes -> double quotes
  # list, tuple -> Array
```

คำสั่ง dump
ใช้ในการเขียนลงไฟล์

```python
with open('ab.json', 'w') as f:
	json.dump(ab,f) # dump ไม่มีตัว s ต่อท้าย เป็นการเขียน object เป็นข้อความ json ลงในไฟล์ fp (แปลง obj เป็น json ก่อน แล้วค่อยเขียนลงไฟล์)
```

```python
import pandas as pd 
# ใช้ Pandas มาช่วยอ่านทำให้กลายเป็น DataFrame
data=pd.read_json('ab.json')
data
```

คีย์ของ dict ใน python สามารถเป็น tuple ได้ แต่ json จะต้องเป็นชนิดตัวแปรเดี่ยวเช่น str, int, float, bool หรือ None เท่านั้น

ตัวอย่างที่ทำไม่ได้

```python
# print(json.dumps({(1,2):1})) error  
# เพราะ json ไม่ supported ค่า key ที่เป็น tuple
# แต่ในกรณีปกติ key สามารถเป็น tuple ได้
```

แต่ถ้าใส่ attribute คือ skipkeys=True จะไม่เกิดข้อผิดพลาด และคีย์เวิร์ดที่เป็น tuple จะถูกข้ามไป

```python
print(json.dumps({(1,2):1, 3:4},skipkeys=True)) # skipkeys=True เป็นการข้าม error output จึงได้แค้ตัวหลังสุด
# จะแปลงได้แค่เทอมหลัง คือ 3:4
# ส่วนเทอมแรกที่เป็น  (1,2):1  จะแปลงไม่ได้ เพราะ key เป็น tuple
```

**ข้อควรระวังของการแปลง object เป็น json**

ใน python มี nan, inf หรือ –inf ซึ่งถ้ามีข้อมูลเหล่านี้ json จะเปลี่ยนเป็น NaN, Infinity, -Infinity

```python
a=float('-inf')
b=float('nan')
c=float('inf')
print([a,b,c]) # ค่าออกมาเป็น string ปกติ
print(json.dumps([a,b,c])) # แปลงค่าให้เพราะเป็น input ที่อยู่ใน dictionary default
```

หากต้องการให้ข้อความ json ที่ได้นั้นแบ่งเป็นหลายบรรทัดแทนที่การเขียนต่อกันเป็นบรรทัดเดียว ให้ใช้ attribute indent โดยระบุจำนวนวรรคในการร่น

```python
print(json.dumps({1:2, 3:[4, [5, {6:7, 8:9}]]}, indent=4)) # ทำให้อ่านง่านด้วย indent โดยเป็นการแสดงค่าให้หลายบรรทัด
```

```python
print(json.dumps({1:2,3:[4,[5,{6:7,8:9}]]})) # ดูยากเพราะไม่ใส่ indent
```

หากต้องการให้มีการจัดลำดับคีย์ใหม่ ให้ใส่ attribute     
  sort_keys=True

```python
dic={'d':1, 'I':2, 'c':3} 
print(json.dumps(dic,sort_keys=True)) # จัดลำดับ key ใหม้ในขั้นตอนการ input Data 
```

ข้อมูลใน dictionary ปกติจะกั้นด้วย : ส่วนใน list จะกั้นด้วย , โดยมีเว้นวรรคตามหนึ่งช่อง อย่างไรก็ตาม สามารถทำการปรับเปลี่ยนโดยใช้ attribute separators

```python
# separators = สร้างตัวกั้นระหว่างข้อมูล
print(json.dumps({1:2,3:[4,[5,{6:7,8:9}]]})) #กั้นด้วย , และ : ที่ีตามด้วย 1 space ตามปกติ
print(json.dumps({1:2,3:[4,[5,{6:7,8:9}]]},separators=[',', ':']))   #กั้นด้วย , และ : ที่ไม่มี space
print(json.dumps({1:2,3:[4,[5,{6:7,8:9}]]},separators=[',  ', ':  '])) #กั้นด้วย , และ : ที่ตามด้วย 2 spaces

# ตัวอย่างที่ดี
print(json.dumps({1:2,3:[4,[5,{6:7,8:9}]]},separators=['+ ', '| '])) #กั้นด้วย + และ | ที่ตามด้วย 1 space
	# Output -> {"1"| 2+ "3"| [4+ [5+ {"6"| 7+ "8"| 9}]]}
```

หาก json พบอักขระที่ไม่ใช่ ASCII จะถูกแปลงเป็นรหัส ซึ่งสามารถแก้ปัญหาได้โดยการใช้ attribute ensure_asci=False

```python
print(json.dumps(['ก', '1'])) # ensuare_ascii=True Default โดยพบค่าที่ไม่อยู่ในตัวอักษรใน ASCII set จะแสดงเป็นค่า error
print(json.dumps(['ก', '1'], ensure_ascii=False)) # แสดงเป็นตัวนั้น ๆ เลย แม้จะไม่พบค่าใน ASCII set
```

การอ่านข้อมูล json

คำสั่ง load
มีไฟล์ json แล้ว ต้องการอ่านขึ้นมาเป็น object ใน python

```python
with open('ab.json') as f:
	obj=json.load(f)
obj # Datatype Object
```

คำสั่ง loads
อ่านข้อความ json ที่เก็บไว้ในตัวแปรหรือเขียนข้อความ json เพื่อแปลงเป็น object

```python
obj=json.loads('["a","b","e"]') #สังเกต loads มีเติม s ด้วย
print(obj)
```

แปลง JSON เป็น python

```python
print(json.loads('["จ",15,3.2,null,false,true,{"4":15}]'))
```

**ข้อควรระวังในการอ่านข้อมูล json**

หากมีข้อมูลที่เป็นรหัส ซึ่งเกิดจากการใช้คำสั่ง dumps โดยไม่ใส่ ensure_ascii=False เวลาใช้ loads ก็จะถูกแปลงกลับเป็นตัวอักษรด้วย

```python
print(json.dumps(['น'])) # อ่าน น ไม่ได้ จึงใช้ค่าอื่นมาแทนค่านี้
```

```python
print(json.loads('["\u0e19"]'))
```

ปกติเลขจำนวนเต็มจะตีความเป็น int แต่สามารถระบุ attribute parse_int เพื่อให้จัดการเป็นอย่างอื่นได้ เช่น เปลี่ยนเลขจำนวนเต็มทั้งหมดให้เป็นเลขทศนิยม

```python
print(json.loads('[1,2.1]',parse_int=float)) # parse_int=float -> เจอ int ให้แปลงเป็น float
```

หากต้องการเปลี่ยนเลขทศนิยมทั้งหมดเป็นจำนวนเต็ม

```python
print(json.loads('[1,2.1]',parse_float=lambda x:int(float(x)))) # เจอ float ให้แปลงเป็น int
```

*************************************************************************************
ปกติ NaN, Infinity, -Infinity ก็จะถูกแปลงเป็น nan, inf, -inf
*************************************************************************************

```python
print(json.loads('[-Infinity,NaN,Infinity]')) # แปลงกลับจาก 68

```

ถ้าต้องการให้ถูกตีความต่างออกไป สามารถกำหนดผ่าน attribute parse_constant ได้ เช่น

```python
print(json.loads('[-Infinity,NaN,Infinity]',parse_constant=str)) 
```
